широта(latitude) земли принимает значения от −90° до 90°. 0° – широта экватора; −90° – широта Южного полюса; 90° – широта Северного полюса
долгота(longitude)-принимает значения от −180° до 180°. Положительные значения соответствуют восточной долготе (сокращённо в.д. или E); отрицательные – западной долготе (сокращённо з.д. или W)._

координаты калифорнии гуглим получаем 
* широта(latitude) 35,1258000
* долгота(longitude) -117,9859000

```python
# longitude - долгота калифорнии отрицательная
# все что + не верно, смотрим
df[df['longitude'] >= 0]
```

```python
# создаем фиктивный признак заполненый нулями,
# делаем маску в которой положительные значения 
# долготы заменяем на еденицу  
df['longitude_outlier'] = 0
df.loc[df['longitude'] >= 0, 'longitude_outlier'] = 1
```

```python
df[df['longitude'] >=0].head()
```

```python
#ошибка сборщика данных потерял минус т.к 
# положительные значения попадают
# под отрицательные кординаты калифорнии
df.loc[df['longitude'] > 0, 'longitude'] * -1
```

```python
#добавляем значения в исходный дата фрейм
df.loc[df['longitude'] > 0,
       'longitude'] = df.loc[df['longitude'] > 0, 'longitude'] * -1
```

```python
#все нулевые значения кординат заменяем на медиану
df.loc[df['longitude'] == 0] = df['longitude'].median()
```

```python
#  с помощю маски проверяем
df[df['longitude']>=0]
```

```python
# latitude калифорнии = 35,1258000 создам маску
df[(df['latitude'] <= 0) | (df['latitude'] > 50)]
```

```python
# снова создаем фиктивный признак
#  и заполняем единицами
df['latitude_outlier'] = 0
df.loc[(df['latitude'] <= 0) | (df['latitude'] > 50), 'latitude_outlier'] = 1
```

```python
df[(df['latitude']<= 0) | (df['latitude'] > 50)]
```
